```python
# Importamos las bibliotecas necesarias para el análisis y visualización de datos
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

# Datos Unidimensionales(Univariados):
Histogramas: Representan la distribución de frecuencia de un conjunto de datos. Útil para entender la forma y la dispersión de los datos.

```python
# Generamos 1000 datos aleatorios de una distribución normal estándar
data = np.random.randn(1000)

# Creamos un histograma de los datos.
# El parámetro 'bins' determina el número de barras en el histograma.
# El parámetro 'edgecolor' define el color de los bordes de las barras.
plt.hist(data, bins=30, edgecolor='black')

# Establecemos el título del gráfico
plt.title('Histograma de Datos Unidimensionales')

# Establecemos la etiqueta del eje x
plt.xlabel('Valor')

# Establecemos la etiqueta del eje y
plt.ylabel('Frecuencia')

# Mostramos el gráfico
plt.show()
```

#Histograma 3D:
Visualiza la distribución de tres variables.

```python
# Generamos 1000 datos aleatorios de una distribución normal estándar para cada eje
x = np.random.randn(1000)
y = np.random.randn(1000)

# La variable z es la suma de x e y
z = x + y

# Creamos una figura con un tamaño específico
fig = plt.figure(figsize=(8, 6))

# Añadimos un subplot en 3D a la figura
ax = fig.add_subplot(111, projection='3d')

# Creamos un gráfico de dispersión en 3D con los datos x, y, z
# El color de cada punto se determina por el valor de z y se utiliza el mapa de colores 'viridis'
ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# Establecemos las etiquetas de los ejes x, y, z
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Establecemos el título del gráfico
ax.set_title('Histograma 3D')

# Mostramos el gráfico
plt.show()
```

#Datos Bidimensionales (Bivariados):
Scatter Plots: Muestra la relación entre dos variables. Útil para identificar patrones o tendencias.

```python
# Generamos 1000 datos aleatorios de una distribución normal estándar para la variable x
x = np.random.randn(1000)

# La variable y es el doble de x más 1, más un ruido aleatorio generado a partir de una distribución normal estándar
y = 2 * x + 1 + 0.5 * np.random.randn(1000)

# Creamos un gráfico de dispersión de los datos x e y
# El color de cada punto es azul
plt.scatter(x, y, color='blue')

# Establecemos el título del gráfico
plt.title('Scatter Plot de Datos Bidimensionales')

# Establecemos las etiquetas de los ejes x e y
plt.xlabel('Variable X')
plt.ylabel('Variable Y')

# Mostramos el gráfico
plt.show()
```

#Gráfico de Dispersion con Tendencia Lineal:

Muestra un gráfico de dispersión con una línea de tendencia lineal.

```python
# Cargamos el conjunto de datos 'tips' de la biblioteca seaborn
data = sns.load_dataset('tips')

# Creamos un gráfico de dispersión con una línea de tendencia
# El parámetro 'x' es la variable independiente, 'total_bill'
# El parámetro 'y' es la variable dependiente, 'tip'
# El parámetro 'data' es el conjunto de datos que estamos utilizando
# 'scatter_kws' es un diccionario de argumentos opcionales para la función plt.scatter()
# 'line_kws' es un diccionario de argumentos opcionales para la función plt.plot() que dibuja la línea de tendencia
sns.regplot(x='total_bill', y='tip', data=data, scatter_kws={'s': 50}, line_kws={'color': 'red'})

# Establecemos el título del gráfico
plt.title('Gráfico de Dispersión con Tendencia Lineal')

# Mostramos el gráfico
plt.show()
```

#Gráfico de Dispersion Hexagonal (Hexbin):

Útil para visualizar la densidad en gráficos de dispersión con muchos puntos.

```python
# Cargar el conjunto de datos 'iris' de seaborn
data = sns.load_dataset('iris')

# Crear el gráfico de dispersión hexagonal usando jointplot
# 'x=data['sepal_length']' y 'y=data['sepal_width']' especifican las variables a utilizar en el eje x y eje y, respectivamente.
# 'kind='hex'' indica que queremos un gráfico de dispersión hexagonal.
g = sns.jointplot(x=data['sepal_length'], y=data['sepal_width'], kind='hex')

# Ajustar la disposición para colocar el título en una posición personalizada
# 'top=0.95' ajusta el espacio en la parte superior del gráfico para dejar espacio para el título.
g.fig.subplots_adjust(top=0.95)

# Agregar título al gráfico
# 'fontsize=16' especifica el tamaño de la fuente del título.
g.fig.suptitle('Gráfico de Dispersión Hexagonal', fontsize=16)

# Mostrar el gráfico
plt.show()
```

#Datos Temporales:
Gráficos de Línea: Útil para mostrar cambios a lo largo del tiempo.

```python
# Creamos un rango de fechas desde '2022-01-01' hasta '2022-01-10'
dates = pd.date_range('2022-01-01', '2022-01-10')

# Generamos valores aleatorios de una distribución normal estándar para cada fecha
# Luego, calculamos la suma acumulativa de estos valores
values = np.random.randn(len(dates)).cumsum()

# Creamos una figura con un tamaño específico
plt.figure(figsize=(13,5))

# Creamos un gráfico de línea con los datos de fechas y valores
# El parámetro 'marker' especifica el tipo de marcador de cada punto en la línea
# El parámetro 'linestyle' especifica el estilo de la línea
# El parámetro 'color' especifica el color de la línea
plt.plot(dates, values, marker='o', linestyle='-', color='green')

# Establecemos el título del gráfico
plt.title('Gráfico de Línea de Datos Temporales')

# Establecemos las etiquetas de los ejes x e y
plt.xlabel('Fecha')
plt.ylabel('Valor')

# Mostramos el gráfico
plt.show()
```

# Diagrama de Caja (Box Plot):
Útil para visualizar la distribución y los valores atípicos en un conjunto de datos.

```python
# Generamos 3 grupos de datos con distribuciones normales diferentes
# Cada grupo tiene 100 datos generados aleatoriamente
data = [np.random.normal(0, 1, 100),  # Grupo 1: media 0, desviación estándar 1
        np.random.normal(0, 1.5, 100),  # Grupo 2: media 0, desviación estándar 1.5
        np.random.normal(0, 0.5, 100)]  # Grupo 3: media 0, desviación estándar 0.5

# Creamos un diagrama de caja con los datos generados
# 'data' son los datos a utilizar, 'labels' son las etiquetas para cada grupo de datos
plt.boxplot(data, labels=['Grupo 1', 'Grupo 2', 'Grupo 3'])  # Crea un diagrama de caja para cada grupo de datos

# Añadimos títulos al gráfico y al eje y
plt.title('Diagrama de Caja')  # Añade un título al gráfico
plt.ylabel('Valores')  # Añade un título al eje y

# Mostramos el gráfico
plt.show()  # Muestra el gráfico
```

#Heatmap (Mapa de Calor):
Útil para visualizar la intensidad de una variable en una matriz.

```python
# Datos para el heatmap (ejemplo)
data = np.random.rand(5, 5)  # Genera datos aleatorios en una matriz 5x5

# Configuración del estilo del heatmap con Seaborn
sns.set_theme()  # Configura el estilo del gráfico con el tema predeterminado de Seaborn

# Creación del heatmap utilizando Seaborn
# 'data' es la matriz de datos, 'cmap' es el esquema de colores, 'annot' muestra los valores en las celdas,
# 'fmt' configura el formato de los valores en las celdas, 'linewidths' establece el ancho de las líneas delimitadoras
sns.heatmap(data, cmap='YlGnBu', annot=True, fmt=".2f", linewidths=.5)  # Crea un heatmap con Seaborn

# Añade un título al heatmap
plt.title('Heatmap con Seaborn')  # Añade un título al gráfico

# Muestra el heatmap
plt.show()  # Muestra el gráfico
```

```python
# Datos para el gráfico de mapa de calor (ejemplo)
data = sns.load_dataset('flights').pivot('month', 'year', 'passengers')
# Carga un conjunto de datos sobre vuelos, realiza un pivote para obtener una matriz de pasajeros organizada por mes y año

# Configuración de la figura con un tamaño específico
plt.figure(figsize=(10, 5))  # Crea una figura de 10x5

# Creación del mapa de calor con Seaborn
# 'data' es la matriz de datos, 'annot' muestra los valores en las celdas,
# 'fmt' configura el formato de los valores en las celdas, 'cmap' es el esquema de colores
sns.heatmap(data, annot=True, fmt='d', cmap='YlGnBu')  # Crea un mapa de calor con Seaborn

# Añade un título al mapa de calor
plt.title('Gráfico de Mapa de Calor con Anotaciones')  # Añade un título al gráfico

# Etiquetas para los ejes x e y
plt.xlabel('Año')  # Etiqueta el eje x
plt.ylabel('Mes')  # Etiqueta el eje y

# Muestra el mapa de calor
plt.show()  # Muestra el gráfico
```

#Gráfico de Radar:
Muestra datos en un formato circular.

```python
# Datos para el gráfico de radar (ajustado)
categories = ['A', 'B', 'C', 'D', 'E']
values1 = [4, 2, 5, 3, 1]
values2 = [3, 4, 2, 5, 1]

# Calcular los ángulos para cada categoría
angles = np.linspace(0, 2 * np.pi, len(categories), endpoint=False)

# Configurar el gráfico de radar
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))  # Crea una figura con un eje polar
ax.plot(angles, values1, marker='o', color='orange', linestyle='-', label='Serie 1')  # Línea y puntos para la Serie 1
ax.fill(angles, values1, color='orange', alpha=0.4)  # Relleno de área para la Serie 1

ax.plot(angles, values2, marker='o', color='blue', linestyle='-', label='Serie 2')  # Línea y puntos para la Serie 2
ax.fill(angles, values2, color='blue', alpha=0.4)  # Relleno de área para la Serie 2

ax.set_title('Gráfico de Radar con Múltiples Series')  # Añade un título al gráfico
ax.set_xticks(angles[:-1])  # Configura las marcas en el eje x (ángulos)
ax.legend()  # Añade una leyenda al gráfico

plt.show()  # Muestra el gráfico
```

#Gráfico de Densidad (Kernel Density Plot):
Muestra la distribución de un conjunto de datos de manera suave.

```python
# Generar datos aleatorios
datos = np.random.randn(1000)  # Genera 1000 datos aleatorios de una distribución normal estándar

# Crear el gráfico de densidad con Seaborn
sns.kdeplot(datos, fill=True)  # Crea un gráfico de densidad con área bajo la curva llenada

# Configurar y mostrar el gráfico
plt.title('Gráfico de Densidad con Seaborn')  # Añade un título al gráfico
plt.xlabel('Valores')  # Etiqueta para el eje x
plt.ylabel('Densidad')  # Etiqueta para el eje y
plt.show()  # Muestra el gráfico
```

```python
# Cargar el conjunto de datos 'iris' de seaborn
data = sns.load_dataset('iris')

# Crear el gráfico de densidad usando jointplot
# 'x=data['sepal_length']' y 'y=data['sepal_width']' especifican las variables a utilizar en el eje x y eje y, respectivamente.
# 'kind='kde'' indica que queremos un gráfico de densidad.
# 'cmap='viridis'' especifica el mapa de colores a utilizar.
g = sns.jointplot(x=data['sepal_length'], y=data['sepal_width'], kind='kde', cmap='viridis')

# Ajustar la disposición para colocar el título en una posición personalizada
# 'top=0.95' ajusta el espacio en la parte superior del gráfico para dejar espacio para el título.
g.fig.subplots_adjust(top=0.95)

# Agregar título al gráfico
# 'fontsize=16' especifica el tamaño de la fuente del título.
g.fig.suptitle('Gráfico de Densidad', fontsize=16)

# Mostrar el gráfico
plt.show()
```

```python
# Cargamos el conjunto de datos 'iris' de seaborn
data = sns.load_dataset('iris')  # Carga el conjunto de datos del iris

# Creamos un gráfico de densidad KDE 2D
# 'sepal_length' y 'sepal_width' son las variables a utilizar
# 'fill=True' significa que el gráfico se rellena, 'cmap' es el mapa de colores
sns.kdeplot(x='sepal_length', y='sepal_width', fill=True, data=data, cmap='Blues')  # Crea un gráfico de densidad KDE 2D

# Añadimos un título al gráfico
plt.title('Gráfico de Densidad KDE 2D')  # Añade un título al gráfico

# Mostramos el gráfico
plt.show()  # Muestra el gráfico
```

#Pair Plot (Gráfico de Pares):

Muestra relaciones bivariadas en un conjunto de datos.

```python
# Importamos las bibliotecas necesarias
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos el conjunto de datos 'iris' de seaborn
data = sns.load_dataset('iris')  # Carga el conjunto de datos del iris

# Creamos un gráfico de pares (pair plot)
# 'data' es el conjunto de datos a utilizar, 'hue' es la variable categórica para colorear los puntos
sns.pairplot(data, hue='species')  # Crea un gráfico de pares para cada par de características en el conjunto de datos

# Mostramos el gráfico
plt.show()  # Muestra el gráfico
```

#Gráfico de Áreas Apiladas:

Muestra la evolución de varias variables a lo largo del tiempo.

```python
# Cargamos el conjunto de datos 'flights' de seaborn
data = sns.load_dataset('flights')  # Carga el conjunto de datos de vuelos

# Creamos una tabla pivotante con los datos
# 'year' es el índice, 'month' son las columnas, 'passengers' son los valores
pivot_data = data.pivot_table(index='year', columns='month', values='passengers')  # Crea una tabla pivotante de pasajeros por año y mes

# Creamos un gráfico de áreas apiladas con los datos de la tabla pivotante
# 'stacked=True' significa que las áreas están apiladas, 'cmap' es el mapa de colores
pivot_data.plot.area(stacked=True, cmap='Blues')  # Crea un gráfico de áreas apiladas de pasajeros por año y mes

# Añadimos un título al gráfico
plt.title('Gráfico de Áreas Apiladas de Pasajeros por Año y Mes')  # Añade un título al gráfico

# Mostramos el gráfico
plt.show()  # Muestra el gráfico
```
