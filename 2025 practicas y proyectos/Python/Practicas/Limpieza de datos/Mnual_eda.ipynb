#Instrucciones Final nulos
Como aun quedan datos nulos voy a verificar si no hay dato con valor de 0
luego si no hay valores con valor 0, reemplazar los nulos con 0, luego de eso pasamos a formato numerico las columnas, eliminando los caracteres y luego sacar la media de cada columna y remmplazar los valores que valgan 0 por la media, esto no es tan necesario con tan pocos datos pero es una demostracion, si es algo para una cantidad superior de datos podemos usar algo mas avanzado como simple imputer.

```python
shiping_fee_zero = df[df["shipping_fee"] == '₹0.00'].shape
item_total_zero = df[df["item_total"] == '₹0.00'].shape

print(f"Nshiping_fee_zero: {shiping_fee_zero[0]}")
print(f"Nitem_total_zero: {item_total_zero[0]}")
```

```python
df['shipping_fee'].fillna('₹0.00', inplace=True)
df['item_total'].fillna('₹0.00', inplace=True)
```

```python
df.isnull().sum()
```

```python
df['shipping_fee'] = df['shipping_fee'].str.replace('₹', '', regex=True).astype(float)
df['item_total'] = df['item_total'].str.replace(r'[₹,]', '', regex=True).astype(float)
```

```python
df.dtypes
```

```python
df.sample(5)
```

#Final Nulos

Como se puede observar luego de todo este proceso ya no tendremos columnas con valores nulos, lo cual nos permitira trabajar con los datos de una mejor manera, ademas de haber dejado cada columna con su correspondiente tipo de dato

```python
item_total_mean = df[df['item_total'] != 0]['item_total'].mean()
shiping_fee_mean = df[df['shipping_fee'] != 0]['shipping_fee'].mean()

print(f"shiping_fee_mean: {shiping_fee_mean}")
print(f"item_total_mean: {item_total_mean}")
```

```python
df[df["shipping_fee"] == 0].shape
```

```python
df['shipping_fee'] = df['shipping_fee'].replace(0, shiping_fee_mean)
df[df["shipping_fee"] == 0].shape
```

```python
df[df["item_total"] == 0].shape
```

```python
df['item_total'] = df['item_total'].replace(0, item_total_mean)
df[df["item_total"] == 0].shape
```
